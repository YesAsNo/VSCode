```python
import numpy as np
import pandas as pd
```

```python
netflix_userbase = pd.read_csv("./Netflix Userbase.csv").to_numpy()
print(np.mean(netflix_userbase[:,6]))
print(np.median(netflix_userbase[:,6]))
print(np.std(netflix_userbase[:,6]))
```

```python
print(np.var(netflix_userbase[:,6]))
print(np.std(netflix_userbase[:,6])**2)
```

```python
print(np.min(netflix_userbase[:,6]))
print(np.max(netflix_userbase[:,6]))
```

```python
print(np.sum(netflix_userbase[:,2]))
print(np.prod(netflix_userbase[:,2]))
```

```python
# Here you get all revenue from "Tablet" users. Mask is used for filtering using true/false.
mask = np.zeros((2500,10),bool)
mask[netflix_userbase[:,8] == "Tablet",:] = True

# Prints amount of tablet users
print("The amount of tablet users =", np.count_nonzero(mask[:,8]))    

# Further filters what is shown from the earlier selection
filtered = netflix_userbase [mask[:,0],2]
print("Summed revenue of tablet users =", np.sum(filtered))
```
